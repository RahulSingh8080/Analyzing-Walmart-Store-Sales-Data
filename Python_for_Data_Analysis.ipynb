2. Setting Up Your Python Environment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Environment is ready")
```

3. Data Manipulation and Cleaning

```python
# Load Data

df = pd.read_csv("sales_data.csv")
```

```python
# Display the first few rows

df.head()
```

```python
# Check for missing values

df.isnull().sum()
```

```python
# Filling the missing values
df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# Filter the data for specific region
Filtered_Data = df[df['Region'] == 'North']

print(Filtered_Data)
```

4. Data Visualization

Basic Visualizations with Matplotlib:

```python
# Sales trend over time

df['Date'] = pd.to_datetime(df['Date'])
plt.plot(df['Date']), df['Sales']
plt.title('Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Bar Chart for regions

region_sales = df.groupby('Region')['Sales'].sum()
region_sales.plot(kind='bar', color='blue')
plt.title('Totoal Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

Advanced Visualizations with Seaborn and Plotly:

```python
import seaborn as sns
import plotly.express as px
```

```python
# histogram for Sales distribution

sns.histplot(df['Sales'], kde=True, color='purple')
plt.title('Sales Distribution')
plt.show()
```

```python
## Interactive scatter plot

fig = px.scatter(df, x = 'Sales', y = 'Profit', color = 'Region', title = 'Sales vs profit')
fig.show()
```

5. Real-World Applications

```python
# Stock market data

stock_df = pd.read_csv('stock_data.csv')
```

```python
stock_df.head()
```

```python
# Calculate daily returns               # (Current-Previous/Previous) * 100

stock_df['Daily_Return'] = stock_df['Close'].pct_change()
```

```python
stock_df.head()
```

```python
# Visualize daily returns

stock_df['Daily_Return'].plot(kind='line', title='Daily stock return' )
plt.show()
```
